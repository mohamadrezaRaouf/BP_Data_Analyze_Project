```python
data_file='C:/Users/Ali/OneDrive/Desktop/xls/xls_files'
dataframecolumn=[]
main_dataframe=pd.DataFrame(columns=['age_group','p_active_emp','p_active_Unemp','p_inactive_edu','p_inactive_homemaker','p_inactive_incomeNojob','p_inactive_other','p_not_stated','gender','county_id','county_name','settelment'])
for file in os.listdir(data_file):
    filename=data_file+'/'+file
    settelment=[]
    if int(filename[len(filename)-5])//2==0:
        for i in range(24):
            settelment.append('روستایی')
    else:
        for i in range(24):
            settelment.append('شهری')
    sheet_names=newData=pd.ExcelFile(filename).sheet_names
    for sheet in sheet_names:
        countyname=sheet[0:len(sheet)-4]
        countyid=sheet[len(sheet)-4:len(sheet)]
        county_id=[]
        county_name=[]
        for i in range(24):
            county_name.append(countyname)
            county_id.append(countyid)
        dataframe=pd.read_excel(filename,sheet)
        dataframe.columns=['age_group','total','total_active','p_active_emp','p_active_Unemp','total_inactive','p_inactive_edu','p_inactive_homemaker','non','p_inactive_incomeNojob','p_inactive_other','p_not_stated']
        for i in range((dataframe.index[dataframe['age_group'] == 'مرد'].tolist())[0]+1):
            dataframe=dataframe.drop(i)
        dataframe=dataframe.drop((dataframe.index[dataframe['age_group'] == 'زن'].tolist())[0])
        gender=[]
        for i in range(12):
            gender.append('مرد')
        for i in range(12):
            gender.append('زن')
        dataframe['gender']=gender
        dataframe=dataframe.drop(columns='non')
        dataframe=dataframe.drop(columns='total')
        dataframe=dataframe.drop(columns='total_active')
        dataframe=dataframe.drop(columns='total_inactive') 
        dataframe['county_id']=county_id
        dataframe['county_name']=county_name
        dataframe['settelment']=settelment
        main_dataframe=pd.concat([main_dataframe,dataframe], ignore_index=True)  
```

```python
dataframemard['+ratio']=dataframemard.iloc[:,[1,3]].sum(axis=1)/dataframemard['total']
dataframemard['-ratio']=dataframemard.iloc[:,[2,4,5,6,7]].sum(axis=1)/dataframemard['total']
dataframemard['diffrences_sort']=(dataframemard['+ratio']-dataframemard['-ratio'])

dataframezan['+ratio']=dataframezan.iloc[:,[1,3,4]].sum(axis=1)/dataframezan['total']
dataframezan['-ratio']=dataframezan.iloc[:,[2,5,6,7]].sum(axis=1)/dataframezan['total']
dataframezan['diffrences_sort']=(dataframezan['+ratio']-dataframezan['-ratio'])
```

مهمترین بخش فاز ۳ این سلول است که در ان شاخص مان را برای موفقیت و نیازمندی در زمینه توسعه اشتغال را بررسی میکنیم

برای بدست اوردن این شاخص ابتدا باید یکسری ستون هارا فاکتور مثبت و یکسری را ستون منفی در نظر بگیریم که این فاکتور ها برای مردان و زنان متفاوت است

برای مردان ما ستون های شاغل و محصل را مثبت در نظر گرفتیم و سپس انهارا جمع و تقسیم بر توتال کرده تا یک عدد بین 0 و 1 به ما دهد و ان عدد را در +ریشیو ذخیره کردیم و همچنین سایر ستون هارا فاکتور منفی درنظر گرفتیم و عملیات مشابهی انجام دادیم

اما برای زنان ما فاکتور مثبت را علاوه بر شاغل و محصل خانه دار نیز در نظر گرفتیم و سایر را فاکتور منفی در نظر گرفتیم و عملیات یکسان مانند مردان برای انها انجام دادیم
